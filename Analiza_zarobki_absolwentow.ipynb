```python
import pandas as pd

major_data = pd.read_csv('graduates-major-data.csv', delimiter = ";", decimal=",")
```

```python
task_1_data = major_data.loc[(major_data['P_N'] >= 30) & (major_data['P_POZIOM'] == "2")]

task_1_data = task_1_data.groupby(['P_KIERUNEK_ID', 'P_KIERUNEK_NAZWA', 'P_NAZWA_UCZELNI'], as_index=False)["P_ME_ZAR_P5"].median()

task_1_data = task_1_data[['P_KIERUNEK_ID', "P_KIERUNEK_NAZWA", "P_NAZWA_UCZELNI", "P_ME_ZAR_P5"]]

task_1_data = task_1_data.sort_values(by='P_ME_ZAR_P5', ascending=False).head(10)

task_1_data["KIERUNEK"] = task_1_data["P_KIERUNEK_NAZWA"].astype("str") + "(" + task_1_data["P_KIERUNEK_ID"].astype("str") + ")"
```

```python
import plotly.express as px
import plotly

graph_data = task_1_data

fig = px.bar(
    graph_data, 
    x = 'KIERUNEK', 
    y = 'P_ME_ZAR_P5',
    color = 'P_NAZWA_UCZELNI',
    
    category_orders=dict(group=list(task_1_data["KIERUNEK"])),
    
    title = "TOP 10 Kierunków z najwyższymi zarobkami (5 lat po ukończeniu studiów II stopnia)",

    labels={
        "P_ME_ZAR_P5": "Średnie zarobki po 5 latach od ukończenia studiów [PLN]",
        "KIERUNEK": "Kierunek",
        "P_NAZWA_UCZELNI": "Szkoła wyższa",
    }
)

fig.update_layout(xaxis_categoryorder = 'total descending')

plotly.offline.plot(fig, filename='TOP_10_Kierunków_zarobki.html')
```

```python
task_3_data = major_data.loc[
    (major_data["P_ROKDYP"] == 2014)
    & (major_data["P_POZIOM"] == "2")
    & (major_data["P_WOJ"] == 14)
]

task_3_data = task_3_data[[
    'P_DZIEDZINA',
    
    'P_ME_ZAR_P1',
    'P_ME_ZAR_P2',
    'P_ME_ZAR_P3',
    'P_ME_ZAR_P4',
    'P_ME_ZAR_P5']]

# task_3_data = task_3_data.set_index('P_DZIEDZINA').T
task_3_data

tmp = pd.DataFrame(columns=['P_DZIEDZINA', "PO_ROKU", 'P_ME_ZAR'])

for index, row in task_3_data.iterrows():
    row_to_append = pd.DataFrame([[str(row['P_DZIEDZINA']), 1, row['P_ME_ZAR_P1']]], columns=['P_DZIEDZINA', "PO_ROKU", 'P_ME_ZAR'])
    tmp = tmp.append(row_to_append, ignore_index=True)
    row_to_append = pd.DataFrame([[str(row['P_DZIEDZINA']), 2, row['P_ME_ZAR_P2']]], columns=['P_DZIEDZINA', "PO_ROKU", 'P_ME_ZAR'])
    tmp = tmp.append(row_to_append, ignore_index=True)
    row_to_append = pd.DataFrame([[str(row['P_DZIEDZINA']), 3, row['P_ME_ZAR_P3']]], columns=['P_DZIEDZINA', "PO_ROKU", 'P_ME_ZAR'])
    tmp = tmp.append(row_to_append, ignore_index=True)
    row_to_append = pd.DataFrame([[str(row['P_DZIEDZINA']), 4, row['P_ME_ZAR_P4']]], columns=['P_DZIEDZINA', "PO_ROKU", 'P_ME_ZAR'])
    tmp = tmp.append(row_to_append, ignore_index=True)
    row_to_append = pd.DataFrame([[str(row['P_DZIEDZINA']), 5, row['P_ME_ZAR_P5']]], columns=['P_DZIEDZINA', "PO_ROKU", 'P_ME_ZAR'])
    tmp = tmp.append(row_to_append, ignore_index=True)
    
task_3_data = tmp[tmp["P_DZIEDZINA"] != "nan"]
```

```python
import plotly.express as px

fig = px.line(
    task_3_data, 
    x = 'PO_ROKU',
    y = 'P_ME_ZAR',
    color = 'P_DZIEDZINA',
    
    title = "Zarobki po ukończeniu studiów wedle dziedziny",
    
    labels={
        "PO_ROKU": "Po n latach od ukończenia",
        "P_ME_ZAR": "Średnie zarobki [PLN]",
        "P_DZIEDZINA": "Dziedzina nauki",
    }
)

plotly.offline.plot(fig, filename='Zarobki_dziedzina.html')
```
